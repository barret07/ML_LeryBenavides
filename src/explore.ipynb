### La unica variable que presente un valor maximo bastante superior a la meida es 'duration'; por lo que calcularemos su IQR.

```python
descr_dur = data_bank_factorized["duration"].describe()

duration_iqr = descr_dur["75%"] - descr_dur["25%"]
uplim_dur = descr_dur["75%"] + 1.5*duration_iqr
lowlim_dur = descr_dur["25%"] - 1.5*duration_iqr
if lowlim_dur < 0:
    lowlim_dur = 0
else: lowlim_dur

print ("1.- Uplimit de 'duration' es:", uplim_dur,"y el Lowlimit es:", lowlim_dur)
```

### Reemplazamos valores superiores a 644.5 por Moda de 'duration'

```python
data_bank_no_outliers = data_bank_factorized.copy()

data_bank_no_outliers.loc [data_bank_no_outliers['duration'] == uplim_dur, 'duration'] = data_bank_no_outliers['duration'].mode()
data_bank_no_outliers.loc [data_bank_no_outliers['emp.var.rate'] <= 0, 'emp.var.rate'] = 0

data_bank_no_outliers.to_excel("/workspaces/ML_LeryBenavides/data/Data_No_Outliers/DATASET_NO_OUTLIERS.xlsx")
```

### Dividimos los datos en Train y Test

```python
from sklearn.model_selection import train_test_split

predictoras = ["job_n", "marital_n", "education_n", "default_n", "housing_n", "loan_n", "contact_n", "month_n", "day_of_week_n", "poutcome_n",
                        "age", "duration", "campaign", "pdays", "previous", "emp.var.rate", "cons.price.idx", "cons.conf.idx", "euribor3m", "nr.employed"]
target =  'y_n'

x = data_bank_no_outliers[predictoras]
y = data_bank_no_outliers[target]

x_train_no_outliers, x_test_no_outliers, y_train, y_test = train_test_split (x, y, test_size = 0.3, random_state = 7)

data_bank_no_outliers_train = pd.concat ([x_train_no_outliers, y_train], axis = 1)
data_bank_no_outliers_train.to_excel ("/workspaces/ML_LeryBenavides/data/Data_No_Outliers/DATASET_TRAIN_NO_OUTLIERS.xlsx")

data_bank_no_outliers_test = pd.concat ([x_test_no_outliers, y_test], axis = 1)
data_bank_no_outliers_test.to_excel ("/workspaces/ML_LeryBenavides/data/Data_No_Outliers/DATASET_TEST_NO_OUTLIERS.xlsx")
```

### ESCALANDO

```python
from sklearn.preprocessing import MinMaxScaler
from pickle import dump

scaler = MinMaxScaler()
scaler.fit(x_train_no_outliers)

x_train_no_outliers_scal = scaler.transform(x_train_no_outliers)
x_train_no_outliers_scal = pd.DataFrame (x_train_no_outliers_scal, index = x_train_no_outliers.index, columns = predictoras)

x_test_no_outliers_scal = scaler.transform(x_test_no_outliers)
x_test_no_outliers_scal = pd.DataFrame (x_test_no_outliers_scal, index = x_test_no_outliers.index, columns = predictoras)

data_bank_no_outliers_train_scal = pd.concat ([x_train_no_outliers_scal, y_train], axis = 1)
data_bank_no_outliers_test_scal = pd.concat ([x_test_no_outliers_scal, y_test], axis = 1)

dump (scaler, open("/workspaces/ML_LeryBenavides/models/No_Outliers_min-Max.sav", "wb"))

data_bank_no_outliers_train_scal
```

## LOGISTIC REGRESSION MODEL

```python
train_data = pd.read_csv ("/workspaces/ML_LeryBenavides/data/processed/clean_train.csv")
test_data = pd.read_csv ("/workspaces/ML_LeryBenavides/data/processed/clean_test.csv")

train_data.head()
```

```python
x_LR_train = train_data.drop(["y_n"], axis = 1)
y_LR_train = train_data ["y_n"]

x_LR_test = test_data.drop(["y_n"], axis = 1)
y_LR_test = test_data ["y_n"]
```

```python
from sklearn.linear_model import LogisticRegression

LR_model = LogisticRegression()

LR_model.fit (x_LR_train, y_LR_train)
```

```python
y_pred = LR_model.predict(x_LR_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy = round((accuracy_score(y_LR_test, y_pred)) * 100, 2)

print ("El accuracy de este modelo de Regresion Lineal es:", accuracy, "%")
```

## MODEL OPTIMIZATION

```python
from sklearn.model_selection import GridSearchCV

hyperparams = {
    "C": [0.001, 0.01, 0.1, 1, 10, 100, 1000],
    "penalty": ["l1", "l2", "elasticnet", None],
    "solver": ["newton-cg", "lbfgs", "liblinear", "sag", "saga"]
}

grid = GridSearchCV(LR_model, hyperparams, scoring = "accuracy", cv = 10)
grid
```

```python
import warnings

with warnings.catch_warnings():
    warnings.simplefilter("ignore")

grid.fit (x_LR_train, y_LR_train)

grid.best_params_
```

```python
LR_best_model = LogisticRegression(C = 0.001, solver = "lbfgs")
LR_best_model.fit (x_LR_train, y_LR_train)
```

```python
y_best_pred = LR_best_model.predict(x_LR_test)
y_best_pred
```

```python
best_accuracy = round((accuracy_score(y_LR_test, y_best_pred)) * 100, 2)

print ("El best_accuracy de este modelo de Regresion Lineal es:", best_accuracy, "%")
```
